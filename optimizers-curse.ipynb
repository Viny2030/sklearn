```python
import polars as pl
import numpy as np

from sklego.datasets import fetch_creditcard

from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import make_scorer, accuracy_score

X, y = fetch_creditcard(return_X_y=True)

settings, cv = 10, 5
params = {
    'random_state': np.arange(settings)
}

grid = GridSearchCV(
    estimator=MLPClassifier(max_iter=25), 
    cv=cv,
    n_jobs=-1,
    param_grid=params, 
    scoring={"acc": make_scorer(accuracy_score)},
    refit="acc"
)
```

```python
%%time 

grid.fit(X, y)
```

```python
pl.DataFrame(grid.cv_results_).plot('param_random_state', 'mean_test_acc', kind='scatter')
```

```python
from sklego.datasets import fetch_creditcard

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, accuracy_score
```

```python
X, y = fetch_creditcard(return_X_y=True)

settings, cv = 10, 5

params = {
    'C': np.linspace(.1, 10.0, settings)
}

grid = GridSearchCV(
    estimator=LogisticRegression(class_weight="balanced"), 
    cv=cv,
    n_jobs=-1,
    param_grid=params, 
    scoring={"acc": make_scorer(accuracy_score)},
    refit="acc"
)

grid.fit(X, y)
```

```python
pl.DataFrame(grid.cv_results_).plot('param_C', 'mean_test_acc', kind='scatter')
```

```python
n_settings = [1, 2, 5, 10, 20, 50]
max_score = []
cv = 5

X, y = fetch_creditcard(return_X_y=True)

for i in n_settings:
    print(i)
    params = {
        'C': np.linspace(.1, 10, i)
    }
    
    grid_base = GridSearchCV(
        estimator=LogisticRegression(class_weight="balanced", max_iter=5),  
        cv=cv,
        n_jobs=-1,
        param_grid=params, 
        scoring={"acc": make_scorer(accuracy_score), 'logloss': 'neg_log_loss'},
        refit=False
    )

    grid_iter = GridSearchCV(
        estimator=LogisticRegression(class_weight="balanced", max_iter=1000), 
        cv=cv,
        n_jobs=-1,
        param_grid=params, 
        scoring={"acc": make_scorer(accuracy_score), 'logloss': 'neg_log_loss'},
        refit=False
    )
    
    max_score.append({
        'i': i,
        'base': grid_base.fit(X, y).cv_results_['mean_test_acc'].max(),
        'maxiter': grid_iter.fit(X, y).cv_results_['mean_test_acc'].max(),
    })
```

```python
pl.DataFrame(max_score).melt(id_vars=['i']).plot(x='i', y='value', color='variable', kind='scatter')
```
